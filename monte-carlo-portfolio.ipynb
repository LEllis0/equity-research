```python
import pandas as pd
import numpy as np
import datetime as dt
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ['AZN', 'GSK', 'HSBA', 'BP', 'SHEL', 'BARC', 'LSEG', 'RIO', 'BATS', 'REL', 'DGE']
tickers = [ticker + '.L' for ticker in tickers]

endDate = dt.datetime.now()
startDate = endDate - dt.timedelta(15*365)
```

```python
def equityData(tickers, startDate, endDate):
    returnsData = yf.download(tickers,start=startDate,end=endDate,interval='1d')
    returnsData = returnsData['Adj Close']
    returnsData = returnsData.pct_change()
    meanReturn = returnsData.mean()
    covMatrix = returnsData.cov()
    return returnsData, meanReturn, covMatrix
```

```python
returnsData, meanReturn, covMatrix = equityData(tickers,startDate,endDate)
returnsData = returnsData.dropna()
portfolioWeight = [1/len(tickers)]*len(tickers)
```
